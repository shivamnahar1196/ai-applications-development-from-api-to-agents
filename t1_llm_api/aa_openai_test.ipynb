```python
import requests
import json


url = "https://api.openai.com/v1/chat/completions"

headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {my_key}"
}

data = {
    "model": "gpt-5.4-mini",
    "messages": [
        {
            "role": "user",
            "content": "Write a one-sentence bedtime story about a unicorn."
        }
    ]
}

response = requests.post(url, headers=headers, json=data)

# Check for success
if response.status_code == 200:
    result = response.json()
    pprint(result)
    # print(result['choices'][0]['message']['content'])
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```

#### *Difference Between the  OpenAI Client &  Raw API*

##### Comparison at a Glance

| Feature | OpenAI Client (client.create) | Raw API (requests.post) |
|---|---|---|
| Ease of Use | High (handles JSON/Headers for you) | Medium (requires manual setup) |
| Maintenance | Updates with new OpenAI features | You must update your JSON structures |
| Error Handling | Auto-retries and specific Error classes | Manual (check status codes) |
| Best For | Daily development, Apps, Agents | Simple scripts, non-Python languages |

Pro-Tip: If you are working in Jupyter Notebooks, stick with the OpenAI Client. It makes your code cleaner and significantly reduces the amount of "boilerplate" code you have to write.


The choice between the OpenAI SDK (Client) and the requests library (API URL) comes down to a trade-off between convenience and control.
###### 1. Using the OpenAI() Client (Recommended)
This is a "wrapper" around the raw API.

* Automatic Parsing: You don't have to deal with JSON headers or manually extracting text from nested dictionaries. You just call completion.choices[0].message.content.
* Built-in Retries: If the API times out or hits a rate limit, the SDK automatically retries the request for you.
* Typing & Validation: If you use an IDE like VS Code, it will provide "Intellisense" (autofill) for parameters like model, temperature, and tools, helping you catch errors before you run the code.
* Streaming Support: Handling real-time streaming (where words appear one by one) is much easier with the SDK’s built-in iterators.

###### 2. Using requests (API URL)
This is a raw HTTP call to OpenAI’s servers.

* Zero Dependencies: You don’t need to install the openai library. This is useful in restricted environments or lightweight "serverless" functions (like AWS Lambda) where keeping the package size small is critical.
* Universal: You can use this same logic in any language (Bash/cURL, JavaScript, Go) without waiting for an official library update.
* Manual Handling: You are responsible for everything—setting the Authorization header, converting your data to JSON, and checking if the response status is 200 OK.
